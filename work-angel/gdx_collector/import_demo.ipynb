```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import csv
import diffops
```

```python
data_folder = "./my_collected_data/"
with open(data_folder + 'pirt_ramp_cart.csv') as csv_file:
    csv_reader = list(csv.reader(csv_file, delimiter=','))
    data_points = np.array(csv_reader[1:])
    t_array = data_points[:,0]
    t_array = t_array.astype(float)
    x_array = data_points[:,1]
    x_array = x_array.astype(float)
    print(np.shape(t_array))
    print(np.shape(x_array))
    
fig0, ax0 = plt.subplots()
plt.grid(alpha=0.5)
ax0.set_ylim([0, None])
ax0.plot(t_array,x_array, 'ro')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title("Position vs Time Graph")
```

```python
dt = t_array[1] - t_array[0]
v_array = np.gradient(x_array, dt)

print(np.shape(v_array))
v_m, v_b = np.polyfit(t_array[1:-1], v_array[1:-1], 1)

fig1, ax1 = plt.subplots()
plt.grid(alpha=0.5)
ax1.set_ylim([0, None])
ax1.plot(t_array, v_array, 'go')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.axline((0, v_b), slope=v_m, alpha=0.5, label=f'$v = {round(v_m, 3)} m/s/s * t + {round(v_b, 3)}m/s$')
plt.legend(loc="best")
plt.title("Velocity vs Time Graph")

print(f'Slope is: {round(v_m, 3)} m/s/s')
print(f'Y-intercept is: {round(v_b, 3)} m/s')
```

```python
a_array = np.gradient(v_array, dt)

print(np.shape(a_array))

fig2, ax2 = plt.subplots()
plt.grid(alpha=0.5)
ax2.set_ylim([0, None])
ax2.plot(t_array, a_array, color='orange')
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s/s)")
plt.title("Acceleration vs Time Graph")
```
