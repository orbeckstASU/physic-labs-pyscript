```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import csv
import diffops
```

```python
with open('pirt_ramp_cart.csv') as csv_file:
    csv_reader = list(csv.reader(csv_file, delimiter=','))
    data_points = np.array(csv_reader[1:])
    t_array = data_points[:,0]
    t_array = t_array.astype(float)
    x_array = data_points[:,1]
    x_array = x_array.astype(float)
    print(np.shape(t_array))
    print(np.shape(x_array))
    
fig0, ax0 = plt.subplots()
ax0.plot(t_array,x_array, 'ro')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
```

```python
dt = t_array[1] - t_array[0]
v_array = np.gradient(x_array, dt)

print(np.shape(v_array))

fig1, ax1 = plt.subplots()
ax1.plot(t_array, v_array, 'go')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
```

```python
a_array = np.gradient(v_array, dt)

print(np.shape(a_array))

fig2, ax2 = plt.subplots()
ax2.plot(t_array, a_array, color='orange')
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s/s)")
```
